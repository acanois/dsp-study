```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sine_sample(x, freq, fs):
    return np.sin((2 * np.pi) * x * (freq / fs))
```

```python
def plot_wave(lambda_func, samplerate, seconds=1, gain=1):
    steps = int(samplerate * seconds)
    x = np.linspace(-1, 1, num=steps)
    y = lambda_func(x)
    plt.xlabel("Samples")
    plt.ylabel("Amplitude")
    plt.plot(y)
```

```python
fs = 44100
nyquist = fs / 2
plot_wave(lambda x: sine_sample(x, nyquist, fs), fs, seconds=1)
```
