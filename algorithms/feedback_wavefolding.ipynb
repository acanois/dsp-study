```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sine_sample(x, freq, fs):
    return np.sin((2 * np.pi) * x * (freq / fs))
```

```python
samplerate = 44100
nyquist = samplerate / 2
seconds = 10
n_samples = samplerate * seconds
phasor = np.linspace(0, n_samples/samplerate, num=1000)
sine = [np.sin(sample) for sample in phasor]

plt.plot(sine)
```

```python
def process(sample, intensity=nyquist):
    tan_sum = np.tanh(sample) 
    return tan_sum + sine_sample(sample, intensity, samplerate)
    
```

```python
def process_block(block):
    return [process(y) * 0.5 for y in np.copy(block)]
    
```

```python
folded = process_block(sine)
```

```python
plt.plot(folded)
```
